```python
#Fahmi Nanda Saputro
#1103200203

import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from ipywidgets import interact
import numpy as np
drive.mount('/content/drive')
#beberapa library untuk memanggil panda, drive, pyplot, widget, dan interact

file_path='/content/drive/MyDrive/dataset/diamonds.csv'
datas=pd.read_csv(file_path)
#pemanggilan path drive

print(datas.head())
```

```python
datas.info()
#mencari metode info dari dataset
```

```python
features=['carat','cut','clarity','depth','table','price','x','y','z']
for feature in features:
  print(f"Jumlah unique value untuk {feature} :{datas[feature].nunique()}")

#melakukan iterasi melalui daftar kolom dalam data yang diolah
```

```python
numerical_features=['carat','depth','table','x','y','z']
categorical_features=['cut','color','clarity']
target=['price']
print("Numerical Features: ",numerical_features)
print("Categorical Features :",categorical_features)
print("Target :",target)

#program untuk mengelompokkan fitur fitur dalam 3 kategori
```

```python
for feature in categorical_features:
  plt.figure(figsize=(7,5))
  sns.countplot(x=feature,data=datas,palette='viridis')
  plt.title(f'Bar chart of {feature}',size=13)
  plt.xlabel(feature,size=11)
  plt.ylabel('Count',size=11)
  plt.show()

#membuat diagram batang tiap kategorikal data
```

```python
for feature in categorical_features:
  plt.figure(figsize=(8,6))
  datas[feature].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=['skyblue','lightgreen','lightcoral','orange','aqua'])
  plt.title(f'Pie chart of {feature}',size=15)
  plt.ylabel('')
  plt.show()

#membuat diagram lingkaran untuk dataset
```

```python
bold_colors=['#FF5733','#C70039','#900C3F','#581845','#4A235A',]
for feature in categorical_features:
  plt.figure(figsize=(8,6))
  datas[feature].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=bold_colors)
  plt.title(f"Pie chart of {feature}",size=15)
  plt.ylabel('')
  plt.show()

#Membuat diagram lingakran menggunakan bold colors
```

```python
for feature in numerical_features:
  plt.figure(figsize=(8,6))
  sns.histplot(datas[feature],kde=True,color='red')
  plt.title(f'Distribution of {feature}',size=13)
  plt.xlabel(feature,size=12)
  plt.ylabel('Frequency',size=12)
  plt.show()

#membuat diagram chart berdasarkan dataset dari diamonds
```

```python
fig, axes = plt.subplots(2,3,figsize=(15,10))
for i ,feature in enumerate(numerical_features):
  sns.histplot(datas[feature],kde=True,color='black',ax=axes[i//3,i%3])
  axes[i//3, i%3].set_title(f'Distribusi {feature}', size=12)
  axes[i//3, i%3].set_xlabel(feature, size=10)
  axes[i//3, i%3].set_ylabel('Frequency', size=10)
plt.tight_layout()
plt.show()

#membuat diagram
```

```python
def plot_histogram(feature,bin_width):
      plt.figure(figsize=(8,6))
      bins=np.arange(min(datas[feature]),max(datas[feature]) + bin_width,bin_width)
      sns.histplot(datas[feature],kde=True,bins=bins,color='blue')
      plt.title(f'Histogram of {feature}',size=12);
      plt.xlabel(feature,size=10)
      plt.ylabel('Frequency',size=10)
      plt.show()
for feature in numerical_features:
  interact(plot_histogram,feature=widgets.fixed(feature),bin_width=(0.1 ,1 ,0.1))

#membuat diagram interaktif berdasarkan dataset yang dibuat
```

```python
fig,axes=plt.subplots(2,3,figsize=(15,8))
for i,feature in enumerate(numerical_features):
  sns.boxplot(x=datas[feature],ax=axes[i//3,i%3],orient='h',color='purple')
  axes[i//3,i%3].set_title(f'Boxplot of {feature}',size=12)
  axes[i//3,i%3].set_ylabel(feature,size=10)

plt.tight_layout()
plt.show()

#membuat boxplot fitur numerik dari dataset yang telah dibuat
```
